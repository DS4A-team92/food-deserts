```python
import pandas as pd
import numpy as np
```

```python
f = open("Underlying Cause of Death, 1999-2019.txt", "r")
print(f.readline())
```

```python
df = pd.read_csv('Underlying Cause of Death, 1999-2019.txt', delimiter = "\t")
df
```

```python
df.columns
```

```python
df2 = df.dropna(subset=['County', 'County Code', 'Cause of death'])
df2
```

```python
df2[df2.County.str.contains('Cook',case=False)]
```

```python
df2 = df2.drop(columns=["Notes"])
df2
```

```python
df.dtypes
```

```python
df2['Crude Rate'] = pd.to_numeric(df['Crude Rate'], errors='coerce') #coverts strings to NaN values
df2
```

```python
df2['Crude Rate'] = df2['Crude Rate'].fillna(0) #coverts NaN to zeros, floating
df2
```

```python
df2.dtypes
```

```python
df2["Crude Rate"] = df2["Crude Rate"].astype("int64")
df2
```

```python
df2.dtypes
```

```python
df2[["Year", "Year Code"]] = df2[["Year", "Year Code"]].apply(pd.to_numeric, errors='coerce', axis=1)
# df2.astype({'Year':'int64'}).dtypes
# cols = ['col1', 'col2', 'col3']
# data[cols] = data[cols].apply(pd.to_numeric, errors='coerce', axis=1)
df2
```

```python
df2.dtypes
```

```python
df2[["County Code","Year","Year Code","Deaths","Population"]] = df2[["County Code","Year","Year Code","Deaths","Population"]].astype("int64")
df2
```

```python
df2.dtypes
```

```python
df2[["County","Cause of death"]] = df2[["County","Cause of death"]].astype("string")
df2
```

```python
df2.dtypes
```

```python
df2 = df2.drop(columns=["Cause of death Code"])
df2
```

```python
df2.dtypes
```

```python
#clean data
df2 = df2[df2.County.str.contains('Cook',case=False)]
df2
```

```python
df2.dtypes
```

```python
df2.to_csv('clean_underlyingcauseofdeath.csv')
```
