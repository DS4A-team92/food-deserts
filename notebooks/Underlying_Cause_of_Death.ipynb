```python
import os
import sys

import seaborn as sns
import matplotlib.pyplot as plt

import numpy as np
import pandas as pd

PROJ_ROOT = os.path.join(os.pardir)
```

```python
file_path = os.path.join(PROJ_ROOT,
                        'data',
                        'raw',
                        'Underlying Cause of Death, 1999-2019.txt')

df = pd.read_csv(file_path, delimiter="\t")
```

```python
df
```

```python
df.shape
```

```python
# Dropping rows without any data
cook_df = df.dropna(subset=['County', 'County Code', 'Cause of death'])
cook_df
```

```python
#Limiting DF to Cook County, Illinois (Chicago Metro Area)
cook_df[cook_df.County.str.contains('Cook',case=False)] 
```

```python
#droping unwanted columns
cook_df = cook_df.drop(columns=["Notes", "Cause of death Code"]) 
cook_df
```

```python
cook_df['Crude Rate'] = pd.to_numeric(df['Crude Rate'], errors='coerce') #coverts strings to NaN values
cook_df['Crude Rate'] = cook_df['Crude Rate'].fillna(0) #coverts NaN to zeros, floating
cook_df
```

```python
# Changing the data types
cook_df[["County Code","Year","Year Code","Deaths","Population", "Crude Rate"]] = cook_df[["County Code","Year","Year Code","Deaths","Population", "Crude Rate"]].astype("int64")
cook_df[["County","Cause of death"]] = cook_df[["County","Cause of death"]].astype("string")
cook_df
```

```python
cook_df.dtypes
```

```python
cook_df.to_csv('../data/processed/clean_underlyingcauseofdeath.csv')
```
