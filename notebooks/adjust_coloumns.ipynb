```python
import os
import sys
import json
import pandas as pd

PROJ_ROOT = os.path.join(os.pardir)
```

```python
clean_data_path = os.path.join(PROJ_ROOT, 
                         'data',
                         'processed',
                         'full_data.csv')
df=pd.read_csv(clean_data_path)
```

```python
df.head(1)
```

```python
json_data_path = os.path.join(PROJ_ROOT, 
                         'data',
                         'interim',
                         'column_names.json')
with open(json_data_path) as json_file:
    column_names = json.load(json_file)
```

```python
df = df.rename(columns = column_names)
```

```python
to_remove=set(df.columns)-set(column_names.values())
```

```python
df=df.drop(to_remove, axis=1)
```

```python
df.head()
```

```python
clean_data_path = os.path.join(PROJ_ROOT, 
                         'data',
                         'processed',
                         'full_data_final.csv')
df.to_csv(clean_data_path,index=False)
```
