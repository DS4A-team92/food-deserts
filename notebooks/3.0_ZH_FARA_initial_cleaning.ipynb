```python
#df=pd.read_csv("data/raw/food_access_file/Food%20Access%20Research%20Atlas.csv")

df=pd.read_csv("../data/raw/food_access_file/Food_Access_Research_Atlas.csv")
```

```python
#Limiting DF to Cook County, Illinois (Chicago Metro Area)
df_cook=df[(df['State']=="Illinois")&(df["County"]=="Cook County")].copy()
```

```python
len(df_cook)
```

```python
#Create a list of all columns with no values
blank_columns=[i for i in df_cook.columns if df_cook[i].isna().sum()==1314]
```

```python
#Drop columns where every value is missing/NaN
df_cook=df_cook.drop(blank_columns,axis=1)
```

```python
df_cook['LA1and10'].value_counts(dropna=False)
```

```python
df_cook.isna().sum()
```

```python
df_cook.info()
```

```python
df_cook=df_cook.fillna(np.nan)
```

```python
#Assess whether missing MedianFamilyIncome values are in low income tracts or non-low income tracts

df_cook[df_cook["MedianFamilyIncome"].isna()]['LowIncomeTracts']
```

```python
#Replace missing MedianFamilyIncome values with the average of the medians of low income tracts 

#low_income_mean=df_cook[df_cook["LowIncomeTracts"]==1]['MedianFamilyIncome'].mean()
#df_cook["MedianFamilyIncome"]=df_cook['MedianFamilyIncome'].replace(np.nan,low_income_mean)

#Using median of low income tracts instead of mean, because the distribution is skewed

low_income_median=df_cook[df_cook["LowIncomeTracts"]==1]['MedianFamilyIncome'].median()
df_cook["MedianFamilyIncome"]=df_cook['MedianFamilyIncome'].replace(np.nan,low_income_median)
```

```python
low_income_median
```

```python
#Dropping columns with only 342 non-null values (which focus on >1 mile access)
df_cook=df_cook.dropna(thresh=935,axis=1)
```

```python
df_cook.info()
```

```python
#Explore how many census tracts with missing values for lasnaphalfshare are in low income census tracts to see if 
#there may be a link between low income census tracks and missing values 
df_cook[df_cook['lasnaphalfshare'].isna()]['LowIncomeTracts'].value_counts(dropna=False)
```

```python
#Explore how many census tracts with missing values for lasnaphalf are in low income census tracts to see if 
#there may be a link between low income census tracks and missing values 
df_cook[df_cook['lasnaphalf'].isna()]['LowIncomeTracts'].value_counts(dropna=False)
```

```python
#Verify that all census tracts with missing values for lasnaphalf also have missing values for lasnaphalfshare
(df_cook[df_cook['lasnaphalf'].isna()]['CensusTract'].value_counts(dropna=False)==df_cook[df_cook['lasnaphalfshare'].isna()]['CensusTract'].value_counts(dropna=False)).value_counts()
```

```python
df_cook[df_cook['lasnaphalf'].isna()]['LILATracts_halfAnd10'].value_counts(dropna=False)
```

```python
#It appears that all missing values for population/share of population variables are NaN if the census tract is not
#tagged as being a census tract with a flag for low access at 0.5 miles for urban areas or 10 miles for rural areas
df_cook[df_cook['LILATracts_halfAnd10']==1]['lasnaphalf'].isna().sum()
```

```python
#It appears that all missing values for population/share of population variables are NaN if the census tract is not
#tagged as being a census tract with a flag for low access at 0.5 miles.
df_cook[df_cook['LATracts_half']==1]['lahisphalf'].isna().sum()
```

```python
#The population living > 0.5 miles from a supermarket is not the same as the census tract population.
#There doesn't seem to be a clear way to impute what lapophalf or lapophalfshare would be in tracts that are not
#flagged as being low access at 0.5 miles. In effect, these variables are treated as being 0 in those census tracts. 
df_cook.iloc[100][['Pop2010','LATracts_half','lapophalf','lapophalfshare']]
```

```python
#Creating a dataframe replacing all NaN values by 0. This may not be the best approach- it may be better to drop 
#the columns with missing values.

#NaN_columns=[i for i in df_cook.columns if df_cook[i].isna().sum()!=0]
df_cook=df_cook.fillna(0)
```
